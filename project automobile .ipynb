```python
import warnings 
warnings.filterwarnings('ignore')
import statsmodels.api as sm
X = df_new.drop('price',axis=True)
y = df_new['price']
xc = sm.add_constant(X)
model = sm.OLS(y,xc).fit()
model.summary()
```

```python
#Backward elimination code
cols=list(X.columns)
pmax=1
while(len(cols)>0):
   p=[]
   X=X[cols]
   xc=sm.add_constant(X)
   model=sm.OLS(y,xc).fit()
   p=pd.Series(model.pvalues.values[1:],index=cols)
   pmax=max(p)
   feature_with_pmax=p.idxmax()
   if(pmax>0.05):
       cols.remove(feature_with_pmax)
   else:
           break
selected_features_BE=cols
print(selected_features_BE)
```

```python
x_update=X[cols]
x_update.head(8)
```

```python
from sklearn.linear_model import Ridge,Lasso,ElasticNet

from sklearn.linear_model import LinearRegression
m1 = LinearRegression()
m2 = Ridge(normalize =True)
m3 = Lasso(normalize = True)
m4 = ElasticNet(normalize=True)
```

```python
from sklearn.model_selection import GridSearchCV
param_rd={'alpha':np.arange(0.01,1,0.01)}
param_Ls={'alpha':np.arange(0.01,1,0.01)}
param_En={'alpha':np.arange(0.01,1,0.01),'l1_ratio':np.arange(0.1,1,0.01)} 
```

```python
Gs_rd=GridSearchCV(m2,param_rd,cv=5,scoring='r2')
Gs_rd.fit(x_update,y)
Gs_rd.best_params_ 
```

```python
Gs_Ls=GridSearchCV(m3,param_Ls,cv=5,scoring='r2')
Gs_Ls.fit(x_update,y)
Gs_Ls.best_params_
```

```python

Gs_En=GridSearchCV(m4,param_En,cv=5,scoring='r2')
Gs_En.fit(x_update,y)
```

```python
Gs_En.best_params_
```

```python
from sklearn.model_selection import KFold
from sklearn import metrics
kf=KFold(n_splits=5,shuffle=True,random_state=0)
for model,name in zip([m1,m2,m3,m4],['Linear_Regression','Ridge','Lasso','ElasticNet']):
    rmse=[]
    for train_idx,test_idx in kf.split(x_update,y):#indexes of test and train willbe stored
        Xtrain,Xtest=x_update.iloc[train_idx,:],x_update.iloc[test_idx,:]
        Ytrain,Ytest=y.iloc[train_idx],y.iloc[test_idx]
        
        model.fit(Xtrain,Ytrain)
        Y_predict=model.predict(Xtest)
        #cm=metrics.confusion_matrix(Ytest,Y_predict)
        mse=metrics.mean_squared_error(Ytest,Y_predict)
        rmse.append(np.sqrt(mse))
    print(model.score(Xtrain,Ytrain))
    print('RMSE scores:%0.03f (+/-%0.5f)[%s]'%(np.mean(rmse),np.std(rmse,ddof=1),name))
```
